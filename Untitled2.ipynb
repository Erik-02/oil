## Start of modeling, column = 'heavy_sour'

#### Seasonal decompose

```python
sdc = seasonal_decompose(df['heavy_sour'], model='multiplicative', period=12)
sdc.plot()
plt.show()
```

#### Plot ACF & PACF

```python
# Stands for Time Series Analysis Plots (TSA Plots)
fig = tsaplots.plot_acf(df['heavy_sour'], lags=60)

plt.xlabel("Lag at k")
plt.ylabel("Correlation coefficient")
plt.show()
```

```python
# Stands for Time Series Analysis Plots (TSA Plots)
fig = tsaplots.plot_pacf(df['heavy_sour'], lags=60)

plt.xlabel("Lag at k")
plt.ylabel("Correlation coefficient")
plt.show()
```

#### Check for stationarity, ADFuller test

```python
#Augmented Dickey–Fuller test:
print('Results of Dickey Fuller Test:')
dftest = adfuller(df['heavy_sour'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```

p-value > 0.05 (significance level), therefore the data is NOT stationary.

since data is not stationary, I will difference the date once, using the 'pandas.diff()' method

```python
df['heavy_sour'] = df['heavy_sour'].diff()
df.dropna(inplace=True)
```

```python
df['heavy_sour'].head()
```

Data has now been differenced once, now we need to check again if data is stationary.

```python
#Augmented Dickey–Fuller test:
print('Results of Dickey Fuller Test:')
dftest = adfuller(df['heavy_sour'], autolag='AIC')

dfoutput = pd.Series(dftest[0:4], index=['Test Statistic','p-value','#Lags Used','Number of Observations Used'])
for key,value in dftest[4].items():
    dfoutput['Critical Value (%s)'%key] = value
    
print(dfoutput)
```

p-value < 0.05 (significance level), therefore, the data is now stationary.

#### Plot data with 12-month moving average

```python
plt.figure(figsize=(20,5))
twelve_months_moving_average = df['heavy_sour'].rolling(window=12).mean()
plt.plot(df['heavy_sour'], color='blue', label='Original')
plt.plot(twelve_months_moving_average, color='red', label='Rolling Mean')
plt.legend(loc='best')
plt.title('12 Months Moving Average on differenced data')
```

```python
# Check data distribution
df['heavy_sour'].hist()
```

Data seems to follow a normal distribution.

#### Differenced data ACF & PACF

```python
# Stands for Time Series Analysis Plots (TSA Plots)
fig = tsaplots.plot_acf(df['heavy_sour'], lags=60)

plt.xlabel("Lag at k")
plt.ylabel("Correlation coefficient")
plt.show()
```

```python
# Stands for Time Series Analysis Plots (TSA Plots)
fig = tsaplots.plot_pacf(df['heavy_sour'], lags=60)

plt.xlabel("Lag at k")
plt.ylabel("Correlation coefficient")
plt.show()
```

```python
# Seasonal - fit stepwise auto-ARIMA
SARIMA_model = pm.auto_arima(df["heavy_sour"], start_p=0, start_q=0,d=1,
                         test='adf', max_p=12, max_q=12, 
                         m=12, #12 is the frequncy of the cycle
                         start_P=0, start_Q=0,
                             max_P=12, max_Q=12,
                         seasonal=True, #set to seasonal
                         D=1, #order of the seasonal differencing
                         trace=False,
                         error_action='ignore',  
                         suppress_warnings=True, 
                         stepwise=True)

print(SARIMA_model.aic())
```

```python
SARIMA_model.summary()
```

```python
SARIMA_model.plot_diagnostics(figsize=(15,12))
plt.show()
```

##### Create training and testing data

```python
df.set_index('period', inplace=True)
```

```python
# Test data is 3 months, since we want to predict the next 3 months
sarima_train, sarima_test = df['heavy_sour'][:-12], df['heavy_sour'][-12:] 
SARIMA_model.fit(sarima_train)
future_periods = sarima_test.size
sarima_forecast = SARIMA_model.predict(n_periods=future_periods)
```

```python
plt.figure(figsize=(20,10))

plt.plot(sarima_train.index[130:], sarima_train.values[130:], color='green', label='train')
plt.plot(sarima_test.index, sarima_test.values, color='red', label='test')
plt.plot(sarima_forecast.index, sarima_forecast.values, color='blue', label='forecast')

# Set plot attributes
plt.xlabel('Months')
plt.ylabel('Average daily imports (measured in million)')
plt.title('Average number of Heavy Sour crude oil imports')
plt.legend()
plt.grid()
plt.show()
```

```python
print(f'Mean absolute percentage error: {mean_absolute_percentage_error(sarima_test, sarima_forecast)}')
print(f'Root mean squared error: {np.sqrt(mean_squared_error(sarima_test, sarima_forecast))}')
print(f'R2 score: {r2_score(sarima_test.values, sarima_forecast.values)}')
```
